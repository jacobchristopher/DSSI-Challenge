```python
# Citation: Code adapted from https://www.kaggle.com/code/plavak10/ecg-heartbeat-classification
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('ecg_dataset/ptbdb_normal.csv')
df2 = pd.read_csv('ecg_dataset/ptbdb_abnormal.csv')
```

```python
dfs = [df1,df2]
for df in dfs:
    df.columns = list(range(len(df.columns)))
```

```python
data = pd.concat(dfs,axis=0).sample(frac=1.0,random_state=1).reset_index(drop=True)
data
```

```python
data.rename(columns={data.columns[-1]:'Target'},inplace=True)
data
```

```python
X = data.drop('Target',axis=1)
y = data['Target']
```

```python
import torch
from pykan.kan import *
torch.set_default_dtype(torch.float64)
# create a KAN: 2D inputs, 1D output, and 5 hidden neurons. cubic spline (k=3), 5 grid intervals (grid=5).
model = KAN(width=[2,5,1], grid=3, k=3, seed=42)
```

```python
from pykan.kan.utils import create_dataset
# create dataset f(x,y) = exp(sin(pi*x)+y^2)
f = lambda x: torch.exp(torch.sin(torch.pi*x[:,[0]]) + x[:,[1]]**2)
dataset = zip(X, y)
dataset['train_input'].shape, dataset['train_label'].shape
```

```python
# plot KAN at initialization
model(dataset['train_input']);
model.plot()
```

```python
# train the model
model.fit(dataset, opt="LBFGS", steps=50, lamb=0.001);
model.plot()
```

```python
model = model.prune()
model.plot()
```

```python
model.fit(dataset, opt="LBFGS", steps=50);
model = model.refine(10)
model.fit(dataset, opt="LBFGS", steps=50);
```

```python
mode = "auto" # "manual"

if mode == "manual":
    # manual mode
    model.fix_symbolic(0,0,0,'sin');
    model.fix_symbolic(0,1,0,'x^2');
    model.fix_symbolic(1,0,0,'exp');
elif mode == "auto":
    # automatic mode
    lib = ['x','x^2','x^3','x^4','exp','log','sqrt','tanh','sin','abs']
    model.auto_symbolic(lib=lib)

model.fit(dataset, opt="LBFGS", steps=50);
```
